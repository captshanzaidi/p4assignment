## 2. Data Loading

We generate synthetic data for demonstration purposes.

```python
# Generate sample data
np.random.seed(42)
n_samples = 100

data = pd.DataFrame({
    'temperature': np.random.normal(23, 2, n_samples),
    'humidity': np.random.normal(65, 10, n_samples),
    'pressure': np.random.normal(1013, 5, n_samples)
})

# Display first 5 rows
data.head()
```

## 3. Statistical Analysis

We compute key statistics for our dataset.

```python
# Calculate statistics
stats = data.describe()
print("Descriptive Statistics:")
print(stats)
```

### 3.1 Correlation Analysis

We examine the relationships between variables.

```python
# Correlation matrix
correlation = data.corr()
correlation
```
